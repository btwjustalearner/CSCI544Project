```python
test1 = pd.read_csv('data/NBtest(CH).csv')
test2 = pd.read_csv('data/NBtest(CH-CHAR).csv')
test3 = pd.read_csv('data/NBtest(CH-CJ).csv')
```

```python
teste = pd.read_csv('data/NBtest(EN).csv')
```

```python
test = test1
test['nbpredchar'] = test2['nbpred']
test['nbpredcj'] = test3['nbpred']
cols = ['nbpred','nbpredchar','nbpredcj']
test['all'] = test[cols].values.tolist() 
```

```python
regdata = teste[['timeseriesforecast','timeserieslower','timeseriesupper','timeseries01','nbpred','groundtruth01']]
```

```python
regdata.dropna(inplace=True)
```

```python
X_train, X_test, y_train, y_test = train_test_split(regdata[['timeseries01','nbpred']],regdata['groundtruth01'], test_size=0.4)
```

```python
clf = LogisticRegression(random_state=0).fit(X_train, y_train)
pred = clf.predict(X_test)
```

```python
f1_score(y_test, pred)
```

```python
def func(x):
    s = list(x)  
    flat_list = [item for sublist in s for item in sublist]
    return max(flat_list, key = flat_list.count)
    
```

```python
df = test.groupby(['date','company'])['all'].apply(func).to_frame()
```

```python
df0 = test.groupby(['date','company'])['groundtruth01'].agg(
    lambda x: pd.Series.mode(x)[0]).to_frame()
```

```python
temp = df0.merge(df, left_index=True, right_index=True)
f1_score(temp['groundtruth01'].round(), temp['all'].round())
```
