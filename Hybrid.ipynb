```python
def getdate(timestr):
    year = '20'+timestr.split('/')[2]
    month = '0'+timestr.split('/')[0]
    month = month[-2:]
    day = '0'+timestr.split('/')[1]
    day = day[-2:]
    return year+'-'+month+'-'+day
```

```python
def func(x):
    s = list(x)  
    flat_list = [item for sublist in s for item in sublist]
    return max(flat_list, key = flat_list.count)
    
```
