```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error, mean_absolute_error
import math
import pmdarima as pm
from statsmodels.tsa.arima_model import ARIMA
```

```python
df = pd.read_csv('AAPL.csv')
```

```python
df['Date']=pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)
```

```python
s = 200
ac_list = []
for row in range(s):   
    train = df[row:(row+30)]
    test = df[(row+30):(row+31)]
    train_log = np.log(train['Close']) 
    test_log = np.log(test['Close'])   
    model = pm.auto_arima(train_log, trace=True, error_action='ignore', suppress_warnings=True)
    model.fit(train_log)
    forecast = model.predict(n_periods=len(test))
    print(test_log[0])
    print(forecast[0])
    print(train_log[-1])
    if (test_log[0]-train_log[-1])*(forecast[0]-train_log[-1]) > 0:
        ac_list.append(1)
    else:
        ac_list.append(0)    
```

```python
sum(ac_list)/s
```

```python
s = 10
for row in range(s):
    train = df[row:(row+30)]
    test = df[(row+30):(row+31)]
    train = train['Close']
    test = test['Close']
    model = ARIMA(train, order=(5,1,1))
    model_fit = model.fit(disp=False)
    forecast, stderr, conf = model_fit.forecast()
    print('Expected: %.3f' % test[0])
    print('Forecast: %.3f' % forecast)
    print('Standard Error: %.3f' % stderr)
    print('95%% Prediction Interval: %.3f to %.3f' % (conf[0][0], conf[0][1]))
```
