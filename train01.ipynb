```python
import pandas as pd
```

```python
data = pd.read_csv('data/traindata0401.csv')
```

```python
data = data[data['nextclose']>0]
```

```python
data.loc[data['nextclose']-data['lastclose'] >= 0, 'groundtruth01'] = int(1)
data.loc[data['nextclose']-data['lastclose'] < 0, 'groundtruth01'] = int(0)
```

```python
data.loc[data['timeseriesforecast']-data['lastclose'] >= 0, 'timeseries01'] = int(1)
data.loc[data['timeseriesforecast']-data['lastclose'] < 0, 'timeseries01'] = int(0)
```

```python
data.to_csv('data/trainCH0407.csv', index=False)
```
